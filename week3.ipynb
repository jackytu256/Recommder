```python
import pandas as pd
import numpy as np
```

```python
documnets = pd.read_csv('../Recommder/Assignment 2.csv')
```

```python
documnets=documnets.set_index('Unnamed: 0').fillna(0).copy()
```

## Part 1. Build and use a very basic profile
    Which document does the simple profile predict user 1 will like best?
    What score does that prediction get?
    How many documents does the model predict U2 will dislike (prediction score that is negative)?

```python
documnets
```

```python
user1_list = []
user2_list = []
for i in documnets.columns:
    if(i != 'User 1' and i != 'User 2'and i != 'num-attr' ):
         print((documnets['User 1'] * documnets[i]).sum(),i)
         print((documnets['User 2'] * documnets[i]).sum(),i)
         user1_list.append((documnets['User 1'] * documnets[i]).sum())
         user2_list.append((documnets['User 2'] * documnets[i]).sum())   
```

```python
df =pd.DataFrame(user1_list,index=documnets.columns[:-3],columns=['user1_profile']).transpose()
df
```

```python
df2 =pd.DataFrame(user2_list,index=documnets.columns[:-3],columns=['user2_profile']).transpose()
df2
```

```python
documnets = documnets.append(df).append(df2)
documnets
```

```python
pd.DataFrame((documnets.ix['doc1',:] * documnets.ix['user1_profile',:]),columns=['user1_profile']).transpose()
```

```python
prediction_user1 = pd.DataFrame()
prediction_user2 = pd.DataFrame()
for i in documnets.index:
    if(i != 'user1_profile' and i != 'user2_profile'and i != 'DF' ):
        print(i)
        prediction_user1 = prediction_user1.append(pd.DataFrame((documnets.ix[i,:] * documnets.ix['user1_profile',:]),columns=[i]).transpose())
        prediction_user2 = prediction_user2.append(pd.DataFrame((documnets.ix[i,:] * documnets.ix['user2_profile',:]),columns=[i]).transpose())
```

```python
prediction_user1
```

```python
test = prediction_user1.transpose().sum()
test.sort(ascending=[1])
test
```

```python
test = (prediction_user2.transpose().sum())
test.sort(ascending=[1])
test
```

## Part 2. Next, let’s treat all articles as having unit weight ...


```python
Part2_documnets = pd.read_csv('../Recommder/Assignment 2.csv').set_index('Unnamed: 0').fillna(0).copy()
```

```python
Part2_documnets
```

```python

weighted_Items = []
for i in Part2_documnets.index:
    print(np.sqrt(Part2_documnets.ix[i,'num-attr'])/Part2_documnets.ix[i,'num-attr'],i)
    weighted_Items.append(np.sqrt(Part2_documnets.ix[i,'num-attr'])/Part2_documnets.ix[i,'num-attr'])
```

```python
aaa =Part2_documnets
aaa
```

```python
for k,i in zip(range(0,len(aaa.index)),aaa.index):
    print(k,i)
    for j,columnName in zip(range(0,len(aaa.ix[i,:])),aaa.columns):
        print(aaa.ix[i,j],j,columnName)
        if(aaa.ix[i,j] == 1 and columnName != "num-attr" and columnName != "User 1" and columnName != "User 2"):
            aaa.ix[i,j] = weighted_Items[k]
```

```python
aaa
```

```python
user1_list = []
user2_list = []
for i in aaa.columns:
    if(i != 'User 1' and i != 'User 2'and i != 'num-attr' ):
         print((aaa['User 1'] * aaa[i]).sum(),i)
         print((aaa['User 2'] * aaa[i]).sum(),i)
         user1_list.append((aaa['User 1'] * aaa[i]).sum())
         user2_list.append((aaa['User 2'] * aaa[i]).sum())   
```

```python
df =pd.DataFrame(user1_list,index=aaa.columns[:-3],columns=['user1_profile']).transpose()
df
```

```python
df2 =pd.DataFrame(user2_list,index=aaa.columns[:-3],columns=['user2_profile']).transpose()
df2
```

```python
aaa = aaa.append(df).append(df2)
aaa
```

```python
prediction_user1 = pd.DataFrame()
prediction_user2 = pd.DataFrame()
for i in aaa.index:
    if(i != 'user1_profile' and i != 'user2_profile'and i != 'DF' ):
        print(i)
        prediction_user1 = prediction_user1.append(pd.DataFrame((aaa.ix[i,:] * aaa.ix['user1_profile',:]),columns=[i]).transpose())
        prediction_user2 = prediction_user2.append(pd.DataFrame((aaa.ix[i,:] * aaa.ix['user2_profile',:]),columns=[i]).transpose())
```
