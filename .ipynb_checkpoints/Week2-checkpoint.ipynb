## % of ratings 4+: Calculate the percentage of ratings for each movie that are 4 or higher. Order with the highest percentage first, and submit the top 5.

```python
top5List = pd.DataFrame()
for i in aiRatings.columns:
    if(i!='User'):
        value = aiRatings[i][aiRatings[i] >=4].count()
        totalCount = aiRatings[i].count()
        top5List.loc[:,i] = pd.Series((value/totalCount)*100)
```

```python
top5List.transpose().sort([0],ascending=[0])
```
